```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, MultiHeadAttention, GlobalAveragePooling1D, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)


X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)

X_train_padded = pad_sequences(X_train_seq, padding='post', maxlen=128)
X_val_padded = pad_sequences(X_val_seq, padding='post', maxlen=128)

label_encoder = LabelEncoder()


y_train_encoded = label_encoder.fit_transform(y_train)
y_val_encoded = label_encoder.transform(y_val)


print(f"Unique labels in y_train after encoding: {np.unique(y_train_encoded)}")
print(f"Unique labels in y_val after encoding: {np.unique(y_val_encoded)}")


vocab_size = len(tokenizer.word_index) + 1


def create_transformer_model(input_dim, output_dim=1):
    inputs = tf.keras.Input(shape=(None,))


    embedding = Embedding(input_dim=input_dim, output_dim=128)(inputs)


    attention = MultiHeadAttention(num_heads=4, key_dim=128)(embedding, embedding)
    attention = tf.keras.layers.LayerNormalization()(attention + embedding)

    pooled = GlobalAveragePooling1D()(attention)


    dense = Dense(64, activation='relu')(pooled)
    dropout = Dropout(0.5)(dense)

    output = Dense(output_dim, activation='sigmoid')(dropout)

    model = tf.keras.Model(inputs=inputs, outputs=output)
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model

transformer_model = create_transformer_model(input_dim=vocab_size)


early_stop = EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)


transformer_history = transformer_model.fit(X_train_padded, y_train_encoded, epochs=10, batch_size=32,
                                            validation_data=(X_val_padded, y_val_encoded), callbacks=[early_stop])


y_pred_transformer = (transformer_model.predict(X_val_padded) > 0.5).astype("int32")

def evaluate_model(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"{model_name} Evaluation:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

evaluate_model(y_val_encoded, y_pred_transformer, 'Transformer')
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc


def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()


def plot_roc_curve(y_true, y_pred, model_name):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='b', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
    plt.title(f'ROC Curve - {model_name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.show()

def plot_metrics_comparison(cnn_metrics, lstm_metrics, transformer_metrics):
    models = ['CNN', 'LSTM', 'Transformer']
    accuracy = [cnn_metrics['accuracy'], lstm_metrics['accuracy'], transformer_metrics['accuracy']]
    precision = [cnn_metrics['precision'], lstm_metrics['precision'], transformer_metrics['precision']]
    recall = [cnn_metrics['recall'], lstm_metrics['recall'], transformer_metrics['recall']]
    f1 = [cnn_metrics['f1'], lstm_metrics['f1'], transformer_metrics['f1']]

    x = np.arange(len(models))
    width = 0.2

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(x - width*1.5, accuracy, width, label='Accuracy')
    ax.bar(x - width*0.5, precision, width, label='Precision')
    ax.bar(x + width*0.5, recall, width, label='Recall')
    ax.bar(x + width*1.5, f1, width, label='F1 Score')

    ax.set_xlabel('Models')
    ax.set_ylabel('Scores')
    ax.set_title('Model Performance Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(models)
    ax.legend()

    plt.show()


def train_and_evaluate(model, X_train, y_train, X_val, y_val, model_name):
    history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stop])
    y_pred = (model.predict(X_val) > 0.5).astype("int32")

    accuracy = accuracy_score(y_val, y_pred)
    precision = precision_score(y_val, y_pred)
    recall = recall_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred)

    print(f"{model_name} - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")

    return history, accuracy, precision, recall, f1, y_pred


cnn_history, cnn_acc, cnn_precision, cnn_recall, cnn_f1, cnn_pred = train_and_evaluate(cnn_model, X_train_padded, y_train_encoded, X_val_padded, y_val_encoded, "CNN")
lstm_history, lstm_acc, lstm_precision, lstm_recall, lstm_f1, lstm_pred = train_and_evaluate(lstm_model, X_train_padded, y_train_encoded, X_val_padded, y_val_encoded, "LSTM")
transformer_history, transformer_acc, transformer_precision, transformer_recall, transformer_f1, transformer_pred = train_and_evaluate(transformer_model, X_train_padded, y_train_encoded, X_val_padded, y_val_encoded, "Transformer")


cnn_metrics = {'accuracy': cnn_acc, 'precision': cnn_precision, 'recall': cnn_recall, 'f1': cnn_f1}
lstm_metrics = {'accuracy': lstm_acc, 'precision': lstm_precision, 'recall': lstm_recall, 'f1': lstm_f1}
transformer_metrics = {'accuracy': transformer_acc, 'precision': transformer_precision, 'recall': transformer_recall, 'f1': transformer_f1}


plot_confusion_matrix(y_val_encoded, cnn_pred, 'CNN')
plot_confusion_matrix(y_val_encoded, lstm_pred, 'LSTM')
plot_confusion_matrix(y_val_encoded, transformer_pred, 'Transformer')


plot_roc_curve(y_val_encoded, cnn_pred, 'CNN')
plot_roc_curve(y_val_encoded, lstm_pred, 'LSTM')
plot_roc_curve(y_val_encoded, transformer_pred, 'Transformer')

plot_metrics_comparison(cnn_metrics, lstm_metrics, transformer_metrics)
```
